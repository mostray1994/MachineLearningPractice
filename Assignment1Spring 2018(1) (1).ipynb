<b>Problem #3-7.</b> Follow steps to answer questions.
> *Use train_test_split() with the option "random_state=0".

1. Fit a linear regression model with RM and LSTAT features only. Find the R-squared metric. 
2. Fit a linear regression model using RM, LSTAT and include the interaction term (RM * LSTAT). How R-squared metric differs from the previous model without interaction term?
3. Fit a linear regression model using LSTAT and include the polynomial term ( $LSTAT^2$).  Find the R-squared metric.
4. Fit linear regression model using LSTAT and include the polynomial term ( $LSTAT^2$ and $LSTAT^3$ ). Find the R-squared metric. 
- How does R-squared metric differ in the previous models ? <br> Comment your observation. 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X = df[['RM','LSTAT']]
Y = dataset.target

X_train, X_test, Y_train, Y_test= train_test_split(X, Y, random_state= 0)

#1 RM and LSTAT only
linreg= LinearRegression()
linreg.fit(X_train, Y_train)
print("The R Square Metric using only RM and LSTAT features is: ",linreg.score(X_test,Y_test))

#2 RM, LSTAT, and (RM*LSTAT)
X1 = df['RM']*df['LSTAT']
X2 = df['RM']
X3 = df['LSTAT']
XC = pd.concat([X2, X3,X1],axis=1)
X_train, X_test, Y_train, Y_test= train_test_split(XC, Y, random_state= 0)
linreg.fit(X_train, Y_train)
print('\n')
print("The R Square Metric using RM, LSTAT, and interaction term(RM*LSTAT) features is: ",linreg.score(X_test,Y_test))

#3 LSTAT and LSTAT^2
XP1 = pd.concat([X3,X3**2],axis=1)
X_train, X_test, Y_train, Y_test= train_test_split(XP1, Y, random_state= 0)
linreg.fit(X_train, Y_train)
print('\n')
print("The R Square Metric using LSTAT and polynomial term(LSTAT*LSTAT) features is: ",linreg.score(X_test,Y_test))

#4
XP2 = pd.concat([X3,X3**2,X3**3],axis=1)
X_train, X_test, Y_train, Y_test= train_test_split(XP2, Y, random_state= 0)
linreg.fit(X_train, Y_train)
print('\n')
print("The R Square Metric using LSTAT and polynomial term(LSTAT^2 and LSTAT^3) features is: ",linreg.score(X_test,Y_test))
```

# ANSWER:

    2) The Interaction term helped relax the additive assumption of the dataset. This means that including both RM and LSTAT as a single feature, the relation between the two made the target prediction more accurate than the model that did not include the interaction term and just had RM and LSTAT as stand alone features.
    
    5) The R square metric improves as more polynomial features are added to the feature list to predict the target. As we could see the scatterplot of LSTAT against the target could be predicted using a polynomial model and adding more polynomial features would help predict the target better to a certain extent.

<b>Problem #3-8.</b> Fit all features (13 features) in the dataset to a multiple linear regression model, and report<br>

<b> Answer the following questions:</b> 

- What does p-value means
- What are the important features? <br>

```python
import statsmodels.formula.api as smf
Xall = df[['target','RM','LSTAT','CRIM','ZN','INDUS','CHAS','NOX','AGE','DIS','RAD','TAX','PTRATIO','B']]

model=smf.ols('target ~ RM+LSTAT+CRIM+ZN+INDUS+CHAS+NOX+AGE+DIS+RAD+TAX+PTRATIO+B', Xall)
Fitting_results=model.fit()
print(Fitting_results.summary().tables[1])
print('p-values are: \n', Fitting_results.pvalues)
```
