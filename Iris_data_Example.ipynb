# Iris classification with KNN 



```python
from sklearn.datasets import load_iris
iris_dataset=load_iris() #creates an object that contains the iris dataset
print('Keys of the iris_dataset is:', iris_dataset.keys())
```

```python
print(iris_dataset.DESCR)
```

```python
iris_dataset['data'].shape
```

```python
import numpy as np
print(iris_dataset['target'][:10]) # printing the first 10 target vaues
print('\n')
print(np.unique(iris_dataset['target'])) # printing the different values in target 
```

```python
#Split your model to training and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=100)
print('size of X_train is:', X_train.shape)
print('size of X_test is:', X_test.shape)
```

```python
#Inpect the data using pandas dataFrame's scatter_plot

%matplotlib inline
import pandas as pd

#use dataframes and label columns with feature names (provided in the dataset)
iris_dataFrame=pd.DataFrame(X_train,columns=iris_dataset.feature_names)

#create a scatter_matrix for the dataframe
sm=pd.plotting.scatter_matrix(iris_dataFrame,c=Y_train, figsize=(15,15))

# seaborn also does pairplots
#import seaborn as sns
#sns.pairplot(iris_dataFrame)
#sns.reset_orig()
```

```python
# Define the machine learning model you will use
from sklearn.neighbors import KNeighborsClassifier 
import numpy as np
k=1 #number of neighbors
knn=KNeighborsClassifier(n_neighbors=k) #initiate class into an object 

knn.fit(X_train, Y_train) # fit the model using training data

accuracy=knn.score(X_test,Y_test) #find the accuracy using test data
print('Accuracy using score method is:', accuracy)

#another way to calculate the accuracy
y_predict=knn.predict(X_test)
print('Accuracy using predict and mean methods', np.mean(y_predict==Y_test))
```
